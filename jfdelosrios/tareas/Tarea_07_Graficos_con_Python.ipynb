```python
import numpy as np
import matplotlib.pyplot as plt
 
n = 256
 
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
 
Y = np.sin(2 * X) 
 
plt.figure(figsize=(10,8))
plt.plot(X, Y + 1, color='blue', alpha=1.00)
 
ax=plt.gca()
ax.fill_between(X, 1, Y+1, color='#9999ff')
 
plt.plot(X, Y - 1, color='blue', alpha=1.00)
ax=plt.gca()
ax.fill_between(X, -1, Y-1, color='#9999ff', where=((Y-1) >= -1))
ax.fill_between(X, -1, Y-1, color='#ff9999',    where=((Y-1) <= -1))
 
plt.xticks([],[])
plt.yticks([],[])
plt.xlim(-np.pi, np.pi)
plt.ylim(-2.5, 2.5)
 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

n = 12

X = np.arange(n)

Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n) 

plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')

plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white') 

for x, y in zip(X, Y1):     

        plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va='bottom') 

plt.ylim(-1.25, +1.25)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
#Z = np.random.uniform(0, 1, 20)
Z = [100/21] * 20
Z[19]=100-np.cumsum(Z)[18]
 
explode = [0.05] * 20
explode[19] = 0.1 
 
cs=[]
for x in range(0, 256, int(256/(len(Z)-1))):
    cs.append('#%02x%02x%02x' % (x, x, x))
 
plt.figure(figsize=(10,10))
 
plt.pie(
    Z, 
    explode = explode, 
    colors=cs, 
    wedgeprops={"edgecolor":"0",'linewidth': 1,'linestyle': 'solid', 'antialiased': True}
    )
 
plt.show()
```
